### Tokeninsing Sentences

```
import nltk
nltk.download('all')
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk import pos_tag
```

```
speech_d = sent_tokenize(speech_d)
speech_d
```

### Word Tokenisation

```
for i in range(len(speech_d)):
  speech_d[i] = word_tokenize(speech_d[i])
print(speech_d)
```

###POS TAGGING

```
for i in range(len(speech_d)):
  speech_d[i] = pos_tag(speech_d[i])
print(speech_d)
```

# Grammar for Verb Phrases

```
grammar = "VP: {<MD>?<VB.*><NP|PP>}"
```

```
results = []
cp = nltk.RegexpParser(grammar)
for sentence in speech_d:
  result = cp.parse(sentence)
  print(result,'\n\n')
  results.append(result)
```
